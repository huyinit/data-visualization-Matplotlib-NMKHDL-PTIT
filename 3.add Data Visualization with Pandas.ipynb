Những âm mưu này không sao, nhưng chúng trông không được bóng bẩy cho lắm. Sử dụng các biểu định kiểu để đặt kiểu thành 'ggplot' và làm lại biểu đồ từ phía trên. Cũng tìm ra cách thêm nhiều thùng vào đó. *

```python
df3['d'].plot.kde()
#Create a kde plot of the 'd' column


```

Tạo một biểu đồ diện tích của tất cả các cột chỉ cho các hàng lên đến 30. (gợi ý: sử dụng .ix).

```python
f = plt.figure()
df3.iloc[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
