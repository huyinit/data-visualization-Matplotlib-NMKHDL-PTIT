Errorbars

Đối với bất kỳ phép đo khoa học nào, việc tính toán chính xác các sai sót gần như quan trọng, nếu không muốn nói là quan trọng hơn việc báo cáo chính xác con số.

Một số giá trị không chính xác một trăm phần trăm và chúng có một số lỗi như giá trị của điện trở. Các giá trị điện trở có dung sai hoặc sai số như 10%. Điều đó có nghĩa là giá trị không chính xác, nhưng nó có thể thay đổi trong phạm vi 10%.

Trong việc trực quan hóa dữ liệu và kết quả, việc hiển thị những lỗi này một cách hiệu quả có thể làm cho một âm mưu truyền tải thông tin đầy đủ hơn nhiều.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

Một thanh lỗi cơ bản có thể được tạo bằng một lệnh gọi hàm Matplotlib:

Ở đây, fmt là một mã định dạng kiểm soát sự xuất hiện của các đường và điểm, đồng thời có cú pháp giống như cách viết tắt được sử dụng trong plt.plot, được nêu trong các Lô đường đơn giản và Lô phân tán đơn giản.

Ngoài các tùy chọn cơ bản này, chức năng thanh lỗi có nhiều tùy chọn để tinh chỉnh kết quả đầu ra. Sử dụng các tùy chọn bổ sung này, bạn có thể dễ dàng tùy chỉnh tính thẩm mỹ của biểu đồ thanh lỗi của mình. Tôi thường thấy hữu ích, đặc biệt là trong các mảnh đất đông đúc, để làm cho các thanh báo lỗi nhẹ hơn chính các điểm:

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

Trong một số tình huống, mong muốn hiển thị các thanh báo lỗi trên số lượng liên tục. Mặc dù Matplotlib không có quy trình tiện lợi tích hợp sẵn cho loại ứng dụng này, nhưng việc kết hợp các nguyên mẫu như plt.plot và plt.fill_between tương đối dễ dàng để có kết quả hữu ích.
