```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
data=np.random.randn(1000)
plt.hist(data)
```

```python
plt.hist(data, edgecolor="black")
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
```

```python
bins=[-3,-2,-1,0,1,2,3]
plt.hist(data, edgecolor="black", bins=bins)
plt.title("My Histograms")
plt.xlabel("Values")
plt.ylabel("Frequencies")
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-3, 2, 1000)
x3 = np.random.normal(2, 1, 1000)
plt.hist(x1, histtype="stepfilled", alpha=0.5, bins=40)
plt.hist(x2, histtype="stepfilled", alpha=0.5, bins=40)
plt.hist(x3, histtype="stepfilled", alpha=0.5, bins=40)
```
